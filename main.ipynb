#Ejercicio 2: ¿Qué velocidad $v$ minimiza la energía empleada en nadar una distancia contracorriente?

Suponga que un pez nada río arriba con velocidad relativa al agua $v$, y que la corriente del río tiene una velocidad de $-V_c$ (el signo negativo indica que la velocidad de la corriente es en dirección opuesta a la del pez). La energía empleada en recorrer una distancia $d$ contracorriente es directamente proporcional al tiempo requerido para recorrer la distancia $d$ y al cubo de la velocidad.

Queremos encontrar la velocidad $v$ que minimiza la energía empleada.

#### Paso 1: Asignar variables
Sea $v$ la velocidad del pez respecto al agua, y $V_c$ la velocidad de la corriente del río. La velocidad neta del pez con respecto a la tierra será $v - V_c$.



#### Paso 2: Función objetivo
La energía $E$ consumida está dada por:
$$
E = k \cdot t \cdot v^3
$$
Donde $k$ es una constante de proporcionalidad, y $t$ es el tiempo que el pez tarda en recorrer la distancia $d$. Dado que el tiempo es la distancia dividida por la velocidad neta, tenemos:
$$
t = \frac{d}{v - V_c}
$$
Sustituyendo esto en la fórmula de la energía:
$$
E(v) = k \cdot \frac{d}{v - V_c} \cdot v^3
$$
Simplificando, obtenemos la siguiente función de energía en términos de $v$:
$$
E(v) = k \cdot d \cdot \frac{v^3}{v - V_c}
$$

#### Paso 3: Encontrar puntos críticos
Para minimizar la energía, derivamos $E(v)$ respecto a $v$ y buscamos los puntos críticos:
$$
E'(v) = k \cdot d \cdot \frac{(3v^2(v - V_c) - v^3)}{(v - V_c)^2}
$$
Igualamos a cero para encontrar el valor de $v$ que minimiza la energía:
$$
3v^2(v - V_c) = v^3
$$
Simplificamos la ecuación:
$$
3v^2(v - V_c) = v^3 \implies 3v - 3V_c = v
$$
$$
2v = 3V_c \implies v = \frac{3}{2} V_c
$$

#### Paso 4: Verificación del mínimo
Para confirmar que este valor de $v$ minimiza la energía, podemos utilizar el criterio de la segunda derivada. Si la segunda derivada de $E(v)$ evaluada en $v = \frac{3}{2} V_c$ es positiva, entonces el punto es un mínimo.

##La velocidad que minimiza la energía es $v = \frac{3}{2} V_c$.


#$2).$Una aplicación: Coste Marginal


En la fabricación y venta de $x$ unidades de un bien de consumo, las funciones de **precio** y **costo** (en dólares) están dadas por:

$$
P(x) = 5.00 - 0.002x \quad \text{(Precio por unidad vendida)}
$$
$$
C(x) = 3.00 + 1.10x \quad \text{(Costo total de producción)}
$$


El ingreso total $I(x)$ está dado por el producto del número de unidades vendidas $x$ y el precio por unidad $P(x)$:

$$
I(x) = x \cdot P(x) = x \cdot (5.00 - 0.002x)
$$

$$
I(x) = 5.00x - 0.002x^2
$$

El **ingreso marginal (IM)** es la derivada del ingreso total $I(x)$ con respecto a $x$:

$$
IM(x) = \frac{d}{dx} \left( 5.00x - 0.002x^2 \right)
$$
$$
IM(x) = 5.00 - 0.004x
$$

El **costo marginal (CM)** es la derivada de la función de costo total $C(x)$ con respecto a $x$:

$$
CM(x) = \frac{d}{dx} \left( 3.00 + 1.10x \right)
$$
$$
CM(x) = 1.10
$$

$$
U(x) = I(x) - C(x)
$$
Sustituyendo las expresiones de $I(x)$ y $C(x)$:
$$
U(x) = (5.00x - 0.002x^2) - (3.00 + 1.10x)
$$

$$
U(x) = -0.002x^2 + 3.90x - 3.00
$$


$$
UM(x) = \frac{d}{dx} \left( -0.002x^2 + 3.90x - 3.00 \right)
$$
$$
UM(x) = -0.004x + 3.90
$$

### Maximización de la utilidad
Para encontrar el nivel de producción que maximiza la utilidad, igualamos la utilidad marginal a cero y resolvemos para $x$:

$$
UM(x) = 0 \quad \Rightarrow \quad -0.004x + 3.90 = 0
$$
$$
0.004x = 3.90
$$
$$
x = \frac{3.90}{0.004} = 975
$$

El nivel de producción que maximiza la utilidad total es **975 unidades**.

### Explicación

##La utilidad es máxima cuando la utilidad marginal es igual a cero. En este caso, el nivel de producción que maximiza la utilidad es 975 unidades.

```python
# cos(t) = t^2
def f1(t):
    return np.cos(t) - t**2

def f1_prime(t):
    return -np.sin(t) - 2*t

# Function for the second equation: 2sin(x) = sqrt(x + 1)
def f2(x):
    return 2 * np.sin(x) - np.sqrt(x + 1)

def f2_prime(x):
    return 2 * np.cos(x) - 1 / (2 * np.sqrt(x + 1))
```

```python
# Plotting and finding initial guesses for f1
x_vals = np.linspace(-2, 2, 400)
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x_vals, f1(x_vals), label='cos(t) - t^2')
plt.axhline(0, color='black', lw=0.5)
plt.title('Graph of cos(t) - t^2')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid(True)
plt.legend()

# Plotting and finding initial guesses for f2
x_vals = np.linspace(-0.5, 5, 400)
plt.subplot(1, 2, 2)
plt.plot(x_vals, f2(x_vals), label='2sin(x) - sqrt(x + 1)')
plt.axhline(0, color='black', lw=0.5)
plt.title('Graph of 2sin(x) - sqrt(x + 1)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
initial_guess_f1 = 0.5 
solution_f1 = newton(f1, initial_guess_f1, f1_prime, tol=1e-4)

initial_guess_f2 = 1.5 
solution_f2 = newton(f2, initial_guess_f2, f2_prime, tol=1e-4)

print("cos(t) = t^2: t ≈", solution_f1)
print("Solution for 2sin(x) = sqrt(x + 1): x ≈", solution_f2)
```

Para encontrar la distancia entre un objeto y el observador hay que utilizar la formula de dos puntos en el plano
$$
d(x) = \sqrt{x^2 + \left(-\frac{x^2}{100} + x + 98\right)^2}
$$
Para encontrar los valores criticos de d(x), se trabajarà con el cuadrado de la distancia para que sea màs manejable que la raìz. Por lo que quedarìa asi:
$$f(x) = x^2 + \left( -\frac{x^2}{100} + x + 98 \right)^2
$$
Se expande
$$\left( -\frac{x^2}{100} + x + 98 \right)^2 = \left( -\frac{x^2}{100} \right)^2 + 2 \left( -\frac{x^2}{100} \right)(x + 98) + (x + 98)^2

Lo podemos simplificar, para que cuando derivemos f(x) quede màs corto

$$f'(x) = \frac{4x^3}{10000} - \frac{6x^2}{100} + \frac{8x}{100} + 196
